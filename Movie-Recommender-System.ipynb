```python
def fetch_director(obj):
    list1 = []
    ## ast_literal is for converting string of list into integer
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            list1.append(i['name'])
            break
    return list1
```

## Removing Spaces between tags

```python
## by lambda x and list comphrension
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

## Creating new column tags 

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords']+ movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head(2)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

### Converting into lowecase

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head(4)
```

## Applying staming technique 
1. for convert (Love Loving Loves To one word Love)


```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

## Finding Similarity Score using Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    ## fetch index of movie
    movie_index = new_df[new_df['title'] == movie].index[0]
    ## dastance from all movies
    distance = similarity[movie_index]
    ## top 5 distances
    movies_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x: x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    return
```

```python
recommend('Batman Begins')
```
